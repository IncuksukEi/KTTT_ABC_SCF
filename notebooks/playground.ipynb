```python
# Cấu hình để vẽ hình ngay trong dòng code
%matplotlib inline 
import sys
import os
import matplotlib.pyplot as plt
import numpy as np

# Thêm đường dẫn thư mục gốc để import được code trong src
sys.path.append(os.path.abspath(os.path.join('..')))

from main import load_config
from src.system_model.channel import ChannelModel
from src.system_model.metrics import SystemMetrics
from src.algorithms.abc_variants import GbestABC

# 1. Chạy nhanh 1 lần mô phỏng
config = load_config('../config.yaml')
M = config['system']['M']
K = config['system']['K']
N = config['system']['N']

print(" Đang chạy mô phỏng G-ABC...")
channel = ChannelModel(M, K, N)
H = channel.generate_rayleigh_channel()
metrics = SystemMetrics(config)
optimizer = GbestABC(config, H, metrics)

best_rate, curve = optimizer.solve()

# 2. Vẽ biểu đồ hội tụ ngay tại đây
plt.figure(figsize=(10, 5))
plt.plot(curve, 'r-o', linewidth=2, label='G-ABC Sum-Rate')
plt.title(f'Kết quả hội tụ: {best_rate:.2f} bps/Hz')
plt.xlabel('Vòng lặp')
plt.ylabel('Sum Rate')
plt.grid(True)
plt.legend()
plt.show()
```
